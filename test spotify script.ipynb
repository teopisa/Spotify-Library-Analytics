```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd
import random

# Carica il dataset
# data = pd.read_csv("nome_file.csv")  # Assicurati di sostituire "nome_file.csv" con il nome reale del tuo file CSV

# Inizializza l'indice per tenere traccia delle righe visualizzate
indice_riga = 0
scelte_file = "scelte.txt"  # File per salvare tutte le scelte dell'utente
top500_file = "top500.txt"  # File per salvare le scelte 'Top 500'
scarti_file = "scarti.txt"  # File per salvare le scelte 'Scarti'

# Carica le scelte precedentemente fatte
try:
    with open(scelte_file, "r") as f:
        scelte = [int(line.strip()) for line in f]
except FileNotFoundError:
    scelte = []

try:
    with open(top500_file, "r") as f:
        top500 = [int(line.strip()) for line in f]
except FileNotFoundError:
    top500 = []

try:
    with open(scarti_file, "r") as f:
        scarti = [int(line.strip()) for line in f]
except FileNotFoundError:
    scarti = []

# Funzione per calcolare la percentuale di canzoni nella Top 500 rispetto a quelle selezionate
def calcola_percentuale_top500():
    totali_selezionati = len(scelte)
    print(totali_selezionati)
    totali_top500 = len(top500)
    print(totali_top500)
    if totali_selezionati > 0:
        percentuale = (totali_top500 / totali_selezionati) * 100
    else:
        percentuale = 0
    return percentuale

# Funzione per visualizzare una nuova riga casuale
def visualizza_riga():
    global indice_riga
    indice_riga = random.choice([i for i in range(len(data)) if i not in scelte])
    label_track_name.config(text=f"Track Name: {data.iloc[indice_riga]['Track Name']}")
    label_artist_name.config(text=f"Artist Name: {data.iloc[indice_riga]['Artist Name(s)']}")
    label_release_date.config(text=f"Release Date: {data.iloc[indice_riga]['Release Date']}")
    label_album_name.config(text=f"Album Name: {data.iloc[indice_riga]['Album Name']}")
    aggiorna_info()

# Funzione per settare il valore della riga corrente
def setta_valore(valore):
    global scelte
    scelte.append(indice_riga)
    data.at[indice_riga, 'Nome_colonna'] = valore
    with open(scelte_file, "a") as f:
        f.write(f"{indice_riga}\n")
    if valore == 1:
        top500.append(indice_riga)
        with open(top500_file, "a") as f:
            f.write(f"{indice_riga}\n")
    else:
        scarti.append(indice_riga)
        with open(scarti_file, "a") as f:
            f.write(f"{indice_riga}\n")
    visualizza_riga()
    # Aggiorna la percentuale di canzoni nella Top 500
    aggiorna_info_percentuale()

# Funzione per aggiornare le informazioni
def aggiorna_info():
    rimanenti = len(data) - len(scelte)
    info_label.config(text=f"Canzoni rimanenti da valutare: {rimanenti}")

# Funzione per aggiornare le informazioni sulla percentuale di canzoni nella Top 500
def aggiorna_info_percentuale():
    percentuale = calcola_percentuale_top500()
    info_percentuale_label.config(text=f"Percentuale canzoni Top 500: {percentuale:.2f}%")

# Crea la finestra principale
root = tk.Tk()
root.title("Gioco di impostazione dati")

# Crea etichette per visualizzare le informazioni della riga corrente
label_track_name = tk.Label(root, text="")
label_track_name.pack()

label_artist_name = tk.Label(root, text="")
label_artist_name.pack()

label_release_date = tk.Label(root, text="")
label_release_date.pack()

label_album_name = tk.Label(root, text="")
label_album_name.pack()

# Crea i pulsanti per impostare il valore a 0 o 1
button_0 = tk.Button(root, text="Scarti", command=lambda: setta_valore(0), font=("Helvetica", 14))
button_0.pack(side=tk.LEFT, padx=10)
button_1 = tk.Button(root, text="Top 500", command=lambda: setta_valore(1), font=("Helvetica", 14))
button_1.pack(side=tk.RIGHT, padx=10)

# Crea un'etichetta per visualizzare le informazioni sulle canzoni rimanenti
info_label = tk.Label(root, text="")
info_label.pack()

# Crea un'etichetta per visualizzare le informazioni sulla percentuale di canzoni nella Top 500
info_percentuale_label = tk.Label(root, text="")
info_percentuale_label.pack()

# Visualizza la prima riga
visualizza_riga()

# Avvia il loop principale
root.mainloop()
```
